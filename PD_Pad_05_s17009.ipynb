````
orders2 = orders
````

````
orders2['tshirt_category'] = orders2['tshirt_category'].astype(str)
````

````
#Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. 
#zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;
````

````
orders2['tshirt_category'] = orders2['tshirt_category'].apply(lambda x: x.replace('Wh ','White '))
orders2['tshirt_category'] = orders2['tshirt_category'].apply(lambda x: x.replace('Tshirt ','T-Shirt '))
orders2['tshirt_category'] = orders2['tshirt_category'].apply(lambda x: x.replace('Bl ','Black '))
````

````
np.unique(orders2['tshirt_category'],return_counts=True) 
````

````
def f(x):
  if len(x.split())==3:
    a = x.split()[0]
    b = x.split()[1]
    c = x.split()[2]
  else:
    a = x.split()[0]
    b = 'Standard'
    c = 'Standard'
  return pd.Series([b, c, a])
````

````
orders3 = orders2
````

````
orders3[['tshirt_type', 'tshirt_size','tshirt_colour']] = orders3['tshirt_category'].apply(f) #Rozdziel kolumnę tshirt category na 3 kolumny
````

````
orders3 = orders3.drop(['tshirt_category'], axis=1)
orders3
````

````
import datetime
````

````
orders4 = orders3[orders3.order_date.between(datetime.datetime(2014, 12, 31), datetime.datetime(2016, 8, 2))] #Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.
orders4
````
